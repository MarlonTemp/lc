```python
class Cell:
    def __init__(self):
        self.numX = 0
        self.numY = 0

    def __repr__(self):
        return "(X: " + str(self.numX) + ", Y: " + str(self.numY) + ")"

    def addCell(self, other):
        self.numX += other.numX
        self.numY += other.numY

    def subCell(self, other):
        self.numX -= other.numX
        self.numY -= other.numY

    def addChar(self, character):
        if character == "X":
            self.numX += 1
        if character == "Y":
            self.numY += 1

    def isBalanced(self):
        return self.numX == self.numY

class Solution:
    def numberOfSubmatrices(self, grid: List[List[str]]) -> int:
        gridHeight = len(grid)
        gridWidth = len(grid[0])

        computed = [[Cell() for i in range(gridWidth)] for i in range(gridHeight)]
        computed[0][0].addChar(grid[0][0])

        for i in range(1, gridWidth): # Adding the first rows values
            computed[0][i].addChar(grid[0][i])
            computed[0][i].addCell(computed[0][i-1])

        for i in range(1, gridHeight): # Adding the first columns values
            computed[i][0].addChar(grid[i][0])
            computed[i][0].addCell(computed[i-1][0])

        for i in range(1, gridHeight): # Looping through each outer array
            for j in range(1, gridWidth): # Looping through inner array
                computed[i][j].addCell(computed[i-1][j])
                computed[i][j].addCell(computed[i][j-1])
                computed[i][j].subCell(computed[i-1][j-1])
                computed[i][j].addChar(grid[i][j])


        total = 0

        for i in range(gridHeight):
            for j in range(gridWidth):
                if (computed[i][j]).isBalanced() and computed[i][j].numX >= 1:
                    total += 1


        return total

```
