## Initial intuition

For this problem, we simply need to keep track of a count, and have a while loop that iterates throught the length of the first word checking if each word's letters match at that index

```python
from typing import List
```

```python
class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
        if len(strs) == 1:
            return strs[0]
        
        word = ""
        firstWord = strs[0]
        while len(word) < len(firstWord):
            curChar = firstWord[len(word)]
            for i in range(1, len(strs)):
                if len(word) >= len(strs[i]):
                    return word
                if strs[i][len(word)] != curChar:
                    return word
                
            word += curChar
        
        return word
        
```

## Result

Not much to say for this one. The solution I made works in 3 seconds, but this easily fluctuates on leetcode

Other solutions are a lot smarter. I looked at a solution by IU Studies (https://leetcode.com/problems/longest-common-prefix/solutions/7500802/clean-simple-beats-easy-to-understand-ef-i2yh/). In their solution, they first sorted the array, then compared the first and last words. This is because, lexicographically ordered, if all the words shared the same prefix, then we would expect the shortest word to be at the front, and the longest word to be at the end. In any other case, this would get messed up, so that either the first or last word would not have the whole of the shared prefix. E.g. lets say we were to have words that share a prefix of pre. For a "valid" set, we could have the data [prep, pre, prefix, premium, pretends, preup], and for "invalid" we could have [prefix, pram, prep, premium, pre, pretends], and the data [prefix, pros, prep, premium, pre, pretends]. (Both are valid, by valid/invalid I mean that they share the whole prefix. Read the examples if still not clear below).

For the valid data, the array would be sorted as so: [pre, prep, preup, prefix, premium, pretends]. We would only have to check if the last and first words share a prefix. Since they do, we know they are valid.

For the first set of invalid data, the array would be sorted *lexicographically* as [pram, pre, prep, prefix, premium, pretends]. We can see that by comparing the first and last words, pram matches pretends only up to pr. This matches every other word's prefix of pr too.

For ths second set of data, we get something similar too: [pre, prep, prefix, premium, pretends, pro]. We can see that the first and last word match up to the prefix "pr".
