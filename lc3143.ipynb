## Initial intuition

What I first noticed is that we only care about the biggest absolute values from each point, as this would define the minimum length the square would have to be. For example, if I required a square to contain a point (0, 10,000), the square would have to be 10,000 in length, despite the other ordinate for the point being 0. Knowing this, we simply need to keep track of the two smallest absolute maximums belonging to a tag. This means if I had three points with the same tag 'a', e.g. (0, 10,000), (5, 10) and (-3, -4), the absolute maximum ordinate from (0, 10,000) would be 10,000. The absolute maximum ordinate from (5, 10) would be 10. The absolute maximum ordinate from (-3, -4) would be 4. Since we would not want to include more than one point of the same tag, we take the point with the second smallest absolute maximum ordinate (which would be (5, 10)), and make the square 1 smaller than the maximum ordinate (so the square would be (-9, -9), (-9, 9), (9, -9) (9, 9)).

To do this, I decided to use a hash map (aka dictionary in python) to store the two smallest maximum absolute ordinates in an array (so given the previous example, I would make a dictionary of ```{a: [4, 10]}``` ). I then iterate through all the tags, finding which has the smallest second smallest, which would be the maximum possible oridnate a point could have in the valid square. I would then count the points, seeing which have ordinates less than or equal to this value.

### Initial attempt

````python
from typing import List

class Solution:
    def maxPointsInsideSquare(self, points: List[List[int]], s: str) -> int:
        tags = {}

        for i in range(len(points)):
            curVal = max(abs(points[i][0]), abs(points[i][1]))
            curTag = s[i]
            
            if curTag in tags: # Checking if dict already set for this element
                if curVal < tags[curTag][0]:
                    tags[curTag][1] = tags[curTag][0]
                    tags[curTag][0] = curVal
                elif curVal < tags[curTag][1]:
                    tags[curTag][1] = curVal
                elif tags[curTag][1] == -1:
                    tags[curTag][1] = curVal

            else:
                tags[curTag] = [curVal, -1]

        minVal = -1

        for i in tags:
            if (tags[i][1] < minVal and tags[i][1] != -1) or minVal == -1:
                minVal = tags[i][1]

        totalPoints = 0

        if minVal == -1:
            return len(points)

        for i in range(len(points)):
            if max(abs(points[i][0]), abs(points[i][1])) < minVal:
                totalPoints += 1

        return totalPoints        
        
````

````python
s = Solution()

print(s.maxPointsInsideSquare([[2,2],[-1,-2],[-4,4],[-3,1],[3,-3]], "abdca"))
````

````python
class Solution:        
    def maxPointsInsideSquare(self, points: List[List[int]], s: str) -> int:
        minLens = {}
        secondMin = float('inf')
        
        for point, char in zip(points, s):
            size = max(abs(point[0]), abs(point[1]))

            if char not in minLens:
                minLens[char] = size
            elif size < minLens[char]:
                secondMin = min(minLens[char], secondMin)
                minLens[char] = size
            else:
                secondMin = min(size, secondMin)
        
        count = 0
        for len in minLens.values():
            if len < secondMin:
                count += 1
        
        return count
````
