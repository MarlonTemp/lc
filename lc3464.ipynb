## Initial intuition

The first thing to point out is that one of the two neighbouring points of any point will have to have the smallest manhattan distance to that point. This led to my first idea of creating some sort of cyclic doubly linked list, and eliminating the point with the closest neighbours one by one, using some data structure to keep track of which point has the smallest neighbours. Eliminating a point would only require the two neighbouring points to be updated, meaning we could sort the points by the minimum distance to another neighbour, and eliminating a point would simply require insertion back into the sorted structure for both the neighbours.

Another main challenge for me is how to elegantly find the neighbours. One way I thought of is to determine whether the angle of a point is greater or lesser than anothers (going clockwise). This means that, if we use the first example given by leetcode, going clockwise, (0, 2) is greater than (0, 0); (2, 2) is greater than (0, 2); (2, 0) is greater than (2, 2); and (0, 0) is greater than (2, 0). This would give us a way to sort the points. I had an interesting thought of using the arctan function (specifically atan2) to determine this value, as it would give an angle from the centre. This would easily allow us to find the ordering of the points on the square. The way the atan2 function works is that, starting from (x, 0), where x is a positive number, the angle is 0. If we continue anticlockwise, the angle increases to a maximum of pi at (-x, 0), then once it continues going around anticlokwise, it goes from -pi to 0 again. I also have to subtract the centre of the square from the points, as to centre the square on the same centre as the unit circle. This would give us a way to find an ordering for the points. This would let us construct our linked list, and then proceed with elimination. 

I realise that this is probably not the intended solution, nor most efficient, but I was curious to whether this solution could work due to the unconventional use of trigonometry.

```python
from typing import List
from math import atan2

class Solution:
    def maxDistance(self, side: int, points: List[List[int]], k: int) -> int:
        pointArr = [Point(i, atan2(i[1] - (side / 2), i[0] - (side / 2))) for i in points]
        pointArr.sort()
        print(pointArr)


class Point:
    def __init__(self, point, angle):
        self.point = point
        self.angle = angle

    def __lt__(self, other):
        return self.angle < other.angle

    def __repr__(self):
        return f"point: ({self.point[0]}, {self.point[1]}), angle: {self.angle}"
        

class ListNode:
    def __init__(self, prevPtr, point, nextPtr):
        self.prevPtr = prevPtr
        self.point = point
        self.val = -1
        self.nextPtr = nextPtr

    def __lt__(self, other):
        return self.val < other.val

    def distToNeightbourLessThan(self, other):
        return min(self.prevPtr.val, self.nextPtr.val)

    def calcVal(self):
        self.val = min(abs(self.point[0] - prevPtr.point[0]) + abs(self.point[1] - prevPtr.point[1]), abs(self.point[0] - nextPtr.point[0]) + abs(self.point[1] - nextPtr.point[1]))

```
