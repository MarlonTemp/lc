```python
from typing import List

class Solution:
    def countSubmatrices(self, grid: List[List[int]], k: int) -> int:
        gridHeight = len(grid)
        gridWidth = len(grid[0])

        computed = [[0 for i in range(gridWidth)] for i in range(gridHeight)]
        computed[0][0] = grid[0][0]

        for i in range(1, gridWidth): # Adding the first rows values
            computed[0][i] = grid[0][i] + computed[0][i-1]

        for i in range(1, gridHeight): # Adding the first columns values
            computed[i][0] = grid[i][0] + computed[i-1][0]

        for i in range(1, gridHeight): # Looping through each outer array
            for j in range(1, gridWidth): # Looping through inner array
                computed[i][j] = computed[i-1][j] + computed[i][j-1] - computed[i-1][j-1] + grid[i][j]

        total = 0

        for i in range(gridHeight):
            for j in range(gridWidth):
                if computed[i][j] <= k:
                    total += 1

        return total
```

```python
testArr = [[7]]
k = 20

x = Solution()
print(x.countSubmatrices(testArr, k))
```
