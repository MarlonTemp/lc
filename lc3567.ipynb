```python
from typing import List

class Solution:
    def minAbsDiff(self, grid: List[List[int]], k: int) -> List[List[int]]:
        gridWidth = len(grid[0]) # n
        gridHeight = len(grid) # m

        result = [[-1 for i in range(gridWidth - k + 1)] for i in range(gridHeight - k + 1)]

        for i in range(gridHeight - k + 1):
            for j in range(gridWidth - k + 1):
                result[i][j] = self.minDiffSubGrid(grid, i, j, k)

        return result

    def minDiffSubGrid(self, grid: List[List[int]], indexX: int, indexY: int, width: int) -> int:
        curMin = -1
        for i in range(indexX, width - 1):
            for j in range(indexY, width - 1):
                for k in range (i + 1, width):
                    for l in range (j + 1, width):
                        curDiff = abs(grid[i][j] - grid[k][l])
                        if curMin == -1 or curDiff < curMin:
                            curMin = curDiff
        return curMin
```
