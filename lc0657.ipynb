```python
class Solution:
    def judgeCircle(self, moves: str) -> bool:
        lrCounter = 0
        udCounter = 0

        for i in moves:
            if i == 'L':
                lrCounter -= 1
            elif i == "R":
                lrCounter += 1

            elif i == 'U':
                udCounter -= 1
            elif i == "D":
                udCounter += 1


        print(udCounter)
        print(lrCounter)
        return lrCounter == 0 and udCounter == 0
```

```python
I = Solution()
print(I.judgeCircle("UUDD"))
```
